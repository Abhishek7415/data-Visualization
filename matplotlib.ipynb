```python
pi = np.pi
x = np.arange(0,7*pi,0.1)
y = np.sin(x)
y2 = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y,'r*')
plt.title('Sine Waveform')
plt.subplot(2,1,2)
plt.plot(x,y2,'b*-')
plt.title('Cosine Waveform')
```
